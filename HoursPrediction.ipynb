```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Read the CSV file
data = pd.read_csv('production.csv')

# Extract the "AMOUNT_TO_CUT" and "AMOUNT_OF_DAYS" columns into arrays
amount_to_cut = data['AMOUNT_TO_CUT'].values.reshape(-1, 1)
amount_of_days = data['AMOUNT_OF_DAYS']

# Print the extracted arrays
# print("AMOUNT_TO_CUT:", amount_to_cut)
# print("AMOUNT_OF_DAYS:", amount_of_days)

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(amount_to_cut, amount_of_days)

# Input the amount to cut
input_amount_to_cut = float(input("Enter the amount to cut: "))

# Predict the number of days based on the input amount
predicted_days = model.predict([[input_amount_to_cut]])

# Print the predicted number of days
print("Predicted number of days to cut {} piece(s): {:.2f}".format(input_amount_to_cut, predicted_days[0]))
```
